```python
import json
import matplotlib.pyplot as plt
```

```python
json_file = "../result.json"
with open(json_file, 'r') as jf:
    results = json.load(jf)
print(results)
```

```python
scores = ['FOM_3x2', 'FOM_DETF_3x2', 'SNR_3x2']
colors = ['red', 'blue', 'green']

for i, score in enumerate(scores):
    fig = plt.figure(figsize=(10, 6))
    for n_bin in range(2, len(results)+2):
        plt.plot(n_bin, results[f'{n_bin}_bins'][score], '.', ms=13, c=colors[i])
    plt.title(score)
    plt.xlabel('Number of bins')
    plt.ylabel(score)
    plt.grid()
    fig.savefig(f'../plots/{score}.png')
```
