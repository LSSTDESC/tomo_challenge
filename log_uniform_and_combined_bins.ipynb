```python
def plot_metrics(bins, scores, title, nrows, ncols, sep_figs = True):
    
    assert len(bins) == len(scores), "The number of bins must be equal to the number of scores (i.e len(bins) == len(scores))"
    
    if sep_figs == False:
        plt.figure(figsize = (10,7))
        plt.plot(bins, [scores[0]['SNR_ww'], scores[1]['SNR_ww'], scores[2]['SNR_ww'], scores[3]['SNR_ww'], scores[4]['SNR_ww']], 'o', label = 'SNR_ww')
        plt.plot(bins, [scores[0]['SNR_gg'], scores[1]['SNR_gg'], scores[2]['SNR_gg'], scores[3]['SNR_gg'], scores[4]['SNR_gg']], 'o', label = 'SNR_gg')
        plt.plot(bins, [scores[0]['SNR_3x2'], scores[1]['SNR_3x2'], scores[2]['SNR_3x2'], scores[3]['SNR_3x2'], scores[4]['SNR_3x2']], 'o', label = 'SNR_3x2')
        plt.plot(bins, [scores[0]['FOM_ww'], scores[1]['FOM_ww'], scores[2]['FOM_ww'], scores[3]['FOM_ww'], scores[4]['FOM_ww']], 'o', label = 'FOM_ww')
        plt.plot(bins, [scores[0]['FOM_gg'], scores[1]['FOM_gg'], scores[2]['FOM_gg'], scores[3]['FOM_gg'], scores[4]['FOM_gg']], 'o', label = 'FOM_gg')
        plt.plot(bins, [scores[0]['FOM_3x2'], scores[1]['FOM_3x2'], scores[2]['FOM_3x2'], scores[3]['FOM_3x2'], scores[4]['FOM_3x2']], 'o', label = 'FOM_3x2')
        plt.plot(bins, [scores[0]['FOM_DETF_ww'], scores[1]['FOM_DETF_ww'], scores[2]['FOM_DETF_ww'], scores[3]['FOM_DETF_ww'], scores[4]['FOM_DETF_ww']], 'o', label = 'FOM_DETF_ww')
        plt.plot(bins, [scores[0]['FOM_DETF_gg'], scores[1]['FOM_DETF_gg'], scores[2]['FOM_DETF_gg'], scores[3]['FOM_DETF_gg'], scores[4]['FOM_DETF_gg']], 'o', label = 'FOM_DETF_gg')
        plt.plot(bins, [scores[0]['FOM_DETF_3x2'], scores[1]['FOM_DETF_3x2'], scores[2]['FOM_DETF_3x2'], scores[3]['FOM_DETF_3x2'], scores[4]['FOM_DETF_3x2']], 'o', label = 'FOM_DETF_3x2')
        plt.xlabel('Number of Bins', fontsize = 14)
        plt.ylabel('Scores', fontsize = 14)
        plt.title(title, fontsize = 16)
        plt.legend()
        
    if sep_figs == True:
        
        metrics = ['SNR_ww', 'SNR_gg', 'SNR_3x2', 'FOM_ww', 'FOM_gg', 'FOM_3x2', 'FOM_DETF_ww', 'FOM_DETF_gg', 'FOM_DETF_3x2']
#         fig, axes = plt.subplots(nrows = nrows, ncols = ncols, figsize = (15,10))
        
        fig = plt.figure(figsize = (ncols*5, nrows*5))
#         fig = plt.figure(figsize = (15,15))
#         plt.xlabel('Number of Bins')
#         plt.axes(frameon=False)
#         plt.xticks([])
#         plt.yticks([])
        plt.axis("off")
        fig.suptitle(title, fontsize = 16)
#         fig.tight_layout()
        fig.subplots_adjust(top = .95)
#         fig.text(0.5, 0.38, 'Number of Bins', fontsize = 14, ha = 'center')
#         fig.text(0.07, 0.5, 'Score', fontsize = 14, va = 'center', rotation = 'vertical')
        
        numplots = nrows*ncols
        
        for i in range(numplots):
            ax = fig.add_subplot(nrows,ncols,i+1)
            ax.plot(bins,[scores[0][metrics[i]], scores[1][metrics[i]], scores[2][metrics[i]], scores[3][metrics[i]], scores[4][metrics[i]]], 'o', label = metrics[i])
            ax.set_xlabel('Number of Bins')
            ax.set_ylabel('Score')
            ax.legend()
        
```
