```python
def plot_onemetric(bins, metric, scores1, bintype1, scores2 = None, bintype2 = None):
    
    y1 = []
    for score in scores1:
        y1.append(score[metric])
    
    plt.figure(figsize = (10,7))
    plt.plot(bins, y1, 'o', label = bintype1)
    
    if scores2 is not None:
        y2 = []
        for score in scores2:
            y2.append(score[metric])
        plt.plot(bins,y2,'o', label = bintype2)
    
    plt.legend()
    plt.xticks(bins, [str(i) for i in bins])
    plt.title(f'{metric} scores')
    plt.xlabel('Number of Bins')
    plt.ylabel('Score')
    plt.savefig('fig2')
```
